```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd

SEED = 42
np.random.seed(SEED)
```

```python
es = 2 * np.random.random((10,10)) - 1
#na = 2 * np.random.random((10,10)) - 1
na = np.array([1,1,0,1,0,1,0,1,1,0])

print(es)
print("---")
print(na)
```

```python
model = Sequential()

model.add(Dense(12, input_dim = 10, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(8,kernel_initializer='uniform',activation='relu'))
model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(es,na,epochs=500,batch_size=20)
```

```python
scores = model.evaluate(es,na)
print("%s: %.2f%%" % (model.metrics_names[1],scores[1]*100))
```

```python
predictions = model.predict(es)
predictions = [int(round(x[0])) for x in predictions]
print(predictions,na)
```
